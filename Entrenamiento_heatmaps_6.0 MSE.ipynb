```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, BatchNormalization, MaxPooling2D, Dropout, Flatten, Dense, Reshape
from tensorflow.keras.regularizers import l2
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Accuracy, MAE
from tensorflow.keras.utils import Sequence
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Define la clase del generador de datos
class DataGenerator(Sequence):
    def __init__(self, imagenes, heatmaps, batch_size=16):
        self.imagenes = imagenes
        self.heatmaps = heatmaps
        self.batch_size = batch_size
        self.indexes = np.arange(len(self.imagenes))

    def __len__(self):
        return int(np.ceil(len(self.imagenes) / self.batch_size))

    def __getitem__(self, index):
        start = index * self.batch_size
        end = (index + 1) * self.batch_size
        batch_x = self.imagenes[start:end]
        batch_y = self.heatmaps[start:end]
        return batch_x, batch_y
```

```python
import tensorflow as tf
from tensorflow.keras.losses import Loss

class CustomLoss(Loss):
    def __init__(self, name='custom_loss'):
        super(CustomLoss, self).__init__(name=name)

    def call(self, y_true, y_pred):
        # y_true y y_pred son tensores de forma (batch_size, 112, 112, 7)
        
        # Calcula la pérdida canal por canal
        channel_losses = []
        for i in range(7):
            # Extrae el canal i de y_true e y_pred
            true_channel = y_true[:, :, :, i]
            pred_channel = y_pred[:, :, :, i]

            # Calcula la pérdida para el canal i (puedes personalizar la pérdida aquí)
            channel_loss = tf.keras.losses.MeanSquaredError()(true_channel, pred_channel)
            
            # Agrega la pérdida del canal i a la lista
            channel_losses.append(channel_loss)
        
        # Devuelve la pérdida total como la media de las pérdidas de cada canal
        return tf.reduce_mean(channel_losses, axis=0)
```

```python
class DataGenerator(Sequence):
    def __init__(self, imagenes, heatmaps, batch_size=16):
        self.imagenes = imagenes
        self.heatmaps = heatmaps
        self.batch_size = batch_size
        self.indexes = np.arange(len(self.imagenes))
        self.some_variable = 0  # Initialize the variable you want to update

    def __len__(self):
        return int(np.ceil(len(self.imagenes) / self.batch_size))

    def __getitem__(self, index):
        start = index * self.batch_size
        end = (index + 1) * self.batch_size
        batch_x = self.imagenes[start:end]
        batch_y = self.heatmaps[start:end]
        return batch_x, batch_y

    def update_var(self, diff):
        # Update the variable using the calculated difference (diff)
        self.some_variable += diff
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Layer

class AdaLoss(Layer):
    def __init__(self):
        super(AdaLoss, self).__init__()

    def call(self, y_true, y_pred):
        return self.loss(y_true, y_pred)

    def loss(self, y_true, y_pred):
        epsilon = 1e-7  # pequeño valor epsilon para estabilidad numérica
        loss = tf.reduce_mean(tf.square(y_true - y_pred), axis=(1, 2, 3)) + epsilon
        return tf.reduce_mean(loss)

class AdaLossCallback(tf.keras.callbacks.Callback):
    def __init__(self):
        super(AdaLossCallback, self).__init__()
        self.loss_list = []
        self.epoch_loss = []
        self.window_size = 3
        self.p = 0.9
        self.previous_loss_var = None  # Initialize here

    def set_datasets(self, train_datagen, val_datagen):
        self.train_datagen = train_datagen
        self.val_datagen = val_datagen

    def on_epoch_end(self, epoch, logs=None):
        epoch_loss_list = [tf.keras.backend.eval(loss) for loss in self.epoch_loss]
        self.epoch_loss.clear()
        loss_mean = np.mean(epoch_loss_list, axis=0)
        self.loss_list.append(loss_mean)

        if len(self.loss_list) > self.window_size:
            del self.loss_list[0]

        if len(self.loss_list) == self.window_size:
            loss_list = np.array(self.loss_list)
            loss_var = np.mean((loss_list - np.mean(loss_list, axis=0)) ** 2, axis=0)

            if self.previous_loss_var is not None:
                diff = 1 - self.previous_loss_var / loss_var
                self.train_datagen.update_var(diff)
                self.val_datagen.update_var(diff)

            self.previous_loss_var = loss_var

            # Imprime la varianza de la pérdida
            print(f'Loss Variance: {loss_var}')

            # Compute new vars
            del self.loss_list[0]  # remove oldest from window

            loss_list = np.array(self.loss_list)
            loss_var = np.mean((loss_list - np.mean(loss_list, axis=0)) ** 2, axis=0)  # calculate loss var in each sample

            if self.previous_loss_var is not None:
                diff = self.p * (1 - self.previous_loss_var / loss_var)
                self.train_datagen.update_var(diff)
                self.val_datagen.update_var(diff)

            self.previous_loss_var = loss_var

```

```python
# Create a model sequential
model = Sequential()

# Base convolucional
model.add(Conv2D(32, (3, 3), padding='same', activation='relu', kernel_regularizer=l2(0.0001), input_shape=(112, 112, 3)))
model.add(BatchNormalization())
model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(256, (3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))

# Regularización L2
model.add(Flatten())
model.add(Dense(112, activation='relu', kernel_regularizer=l2(0.001)))
model.add(BatchNormalization())
model.add(Dense(112 * 112 * 7, activation='sigmoid'))
model.add(Dropout(0.2))
model.add(Reshape((112, 112, 7, 1)))

opt = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=opt, loss=AdaLoss())

# Create instances of the data generator for training, validation, and test
train_generator = DataGenerator(imagenes_train, heatmaps_train, batch_size=16)
val_generator = DataGenerator(imagenes_val, heatmaps_val, batch_size=16)
test_generator = DataGenerator(imagenes_test, heatmaps_test, batch_size=16)

# Initialize AdaLossCallback and set datasets
ada_loss_callback = AdaLossCallback()
ada_loss_callback.set_datasets(train_generator, val_generator)

# Train the model with the AdaLossCallback
history = model.fit(train_generator, epochs=3, validation_data=val_generator, callbacks=[ada_loss_callback])

# Evaluate the model on test data using the generator
test_loss = model.evaluate(test_generator)
print(f'Loss on test data: {test_loss}')
```

```python
import cv2

# Selecciona una imagen de imagenes_test
indice_imagen_test = 10  # Cambia esto al índice de la imagen que deseas usar
img_array = imagenes_test[indice_imagen_test]

# Añade una dimensión extra para el lote
img_array = np.expand_dims(img_array, axis=0)

# Realiza la predicción
heatmap_pred = model.predict(img_array)

# Inicializa la máscara final
final_mask = np.zeros_like(heatmap_pred[0, :, :, 0, 0])

# Itera sobre los canales y agrega las máscaras individuales a la máscara final
for canal in range(heatmap_pred.shape[-2]):
    # Encuentra la posición del valor máximo en el canal predicho
    max_position = np.unravel_index(np.argmax(heatmap_pred[0, :, :, canal, 0]), heatmap_pred.shape[1:3])

    # Crea una máscara binaria
    mask = np.zeros_like(heatmap_pred[0, :, :, canal, 0])
    mask[max_position] = 1

    # Agrega la máscara individual a la máscara final
    final_mask += mask

# Aplica un umbral a la máscara final
threshold = 0.5 * final_mask.max()  # Ajusta el umbral según sea necesario
final_mask = (final_mask > threshold).astype(np.uint8)

plt.figure(figsize=(12, 4))

# Muestra la imagen original
plt.subplot(1, 3, 1)
plt.imshow(img_array[0])
plt.title('Imagen Original')

# Muestra la máscara final
plt.subplot(1, 3, 2)
plt.imshow(final_mask, cmap='gray')
plt.title('Máscara Final')
plt.colorbar()

# Muestra las superposiciones entre la imagen original y la máscara final
plt.subplot(1, 3, 3)
plt.imshow(img_array[0])
plt.imshow(final_mask, alpha=0.3, cmap='jet')  # Ajusta la transparencia según sea necesario
plt.title('Superposición')

plt.show()
```
