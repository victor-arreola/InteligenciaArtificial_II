```python
from tensorflow.keras.utils import Sequence

# Define la clase del generador de datos
class DataGenerator(Sequence):
    def __init__(self, imagenes, heatmaps, batch_size=16):
        self.imagenes = imagenes
        self.heatmaps = heatmaps
        self.batch_size = batch_size
        self.indexes = np.arange(len(self.imagenes))

    def __len__(self):
        return int(np.ceil(len(self.imagenes) / self.batch_size))

    def __getitem__(self, index):
        start = index * self.batch_size
        end = (index + 1) * self.batch_size
        batch_x = self.imagenes[start:end]
        batch_y = self.heatmaps[start:end]
        return batch_x, batch_y
```

```python
import tensorflow as tf
import numpy as np

# Callback de AdaLoss
class AdaLossCallback(tf.keras.callbacks.Callback):

    def __init__(self):
        super(AdaLossCallback, self).__init__()
        self.loss_list = []  # Lista de pérdidas por época
        self.epoch_loss = []  # Pérdida acumulada por lotes en cada época (n_batches, n_samples/batch, landmarks)
        self.window_size = 3  # Tamaño de la ventana para calcular la varianza móvil
        self.p = 0.9  # Factor de actualización de la varianza

    def set_datasets(self, train_datagen, val_datagen):
        self.previous_loss_var = None
        self.train_datagen = train_datagen
        self.val_datagen = val_datagen

    def loss(self, y_true, y_pred):
        # Función de pérdida MSE (Mean Squared Error)
        loss = tf.reduce_mean((y_true - y_pred) ** 2, axis=(1, 2))  # mse, (samples, landmarks)
        self.epoch_loss.append(loss)
        return tf.reduce_mean(loss)

    def on_epoch_end(self, epoch, logs=None):
        # Calcular la media de las pérdidas acumuladas en cada época
        epoch_loss_list = np.array(self.epoch_loss)
        self.epoch_loss.clear()
        loss_mean = epoch_loss_list.mean(axis=(0, 1))  # lista n_landmarks
        self.loss_list.append(loss_mean)  # lista (épocas, n_landmarks)

        # Calcular nuevas variables
        if len(self.loss_list) > self.window_size:
            del self.loss_list[0]  # eliminar la más antigua de la ventana
        if len(self.loss_list) == self.window_size:
            loss_list = np.array(self.loss_list)
            loss_var = ((loss_list - loss_list.mean(axis=0)) ** 2).mean(axis=0)  # calcular la varianza de pérdida en cada landmark

            if self.previous_loss_var is not None:
                diff = self.p * (1 - self.previous_loss_var / loss_var)
                self.train_datagen.update_var(diff)
                self.val_datagen.update_var(diff)
            self.previous_loss_var = loss_var
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.regularizers import L2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPool2D, Dropout, UpSampling2D, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import GlobalAveragePooling2D, Reshape, Dense, AveragePooling2D, BatchNormalization, Activation, Concatenate

def deeplab_plus(input_size=(112, 112, 3), n_classes=7):
    def SqueezeAndExcite(inputs, ratio=8):
        # Capa de Squeeze-and-Excite para realzar características relevantes
        init = inputs
        filters = init.shape[-1]
        se_shape = (1, 1, filters)

        se = GlobalAveragePooling2D()(init)
        se = Reshape(se_shape)(se)
        se = Dense(filters // ratio, activation='relu', kernel_initializer='he_normal', use_bias=False)(se)
        se = Dense(filters, activation='sigmoid', kernel_initializer='he_normal', use_bias=False)(se)
        x = init * se
        return x

    def ASPP(inputs):
        """ Módulo de Atrous Spatial Pyramid Pooling (ASPP) """
        # Red de convolución atrous con diferentes tasas de dilatación
        shape = inputs.shape
        
        # Image Pooling
        y1 = AveragePooling2D(pool_size=(shape[1], shape[2]))(inputs)
        y1 = Conv2D(256, 1, padding="same", use_bias=False)(y1)
        y1 = BatchNormalization()(y1)
        y1 = Activation("relu")(y1)
        y1 = UpSampling2D((shape[1], shape[2]), interpolation="bilinear")(y1)

        # Convolución 1x1
        y2 = Conv2D(256, 1, padding="same", use_bias=False)(inputs)
        y2 = BatchNormalization()(y2)
        y2 = Activation("relu")(y2)

        # Convolución 3x3 con tasa de dilatación 6
        y3 = Conv2D(256, 3, padding="same", use_bias=False, dilation_rate=6)(inputs)
        y3 = BatchNormalization()(y3)
        y3 = Activation("relu")(y3)

        # Convolución 3x3 con tasa de dilatación 12
        y4 = Conv2D(256, 3, padding="same", use_bias=False, dilation_rate=12)(inputs)
        y4 = BatchNormalization()(y4)
        y4 = Activation("relu")(y4)

        # Convolución 3x3 con tasa de dilatación 18
        y5 = Conv2D(256, 3, padding="same", use_bias=False, dilation_rate=18)(inputs)
        y5 = BatchNormalization()(y5)
        y5 = Activation("relu")(y5)

        y = Concatenate()([y1, y2, y3, y4, y5])
        y = Conv2D(256, 1, padding="same", use_bias=False)(y)
        y = BatchNormalization()(y)
        y = Activation("relu")(y)

        return y

    """ Definición de la arquitectura DeepLab+ """
    # Capa de entrada
    inputs = Input(input_size)

    """ Encoder """
    # Utilizando ResNet50 preentrenada en ImageNet como codificador
    encoder = ResNet50(weights="imagenet", include_top=False, input_tensor=inputs)

    # Obtener características de la capa conv4_block6_out
    image_features = encoder.get_layer("conv4_block6_out").output
    x_a = ASPP(image_features)
    x_a = UpSampling2D((4, 4), interpolation="bilinear")(x_a)

    # Obtener características de la capa conv2_block2_out
    x_b = encoder.get_layer("conv2_block2_out").output
    x_b = Conv2D(filters=48, kernel_size=1, padding='same', use_bias=False)(x_b)
    x_b = BatchNormalization()(x_b)
    x_b = Activation('relu')(x_b)

    # Concatenar características de ASPP y conv2_block2_out
    x = Concatenate()([x_a, x_b])
    x = SqueezeAndExcite(x)

    # Más capas convolucionales
    x = Conv2D(filters=256, kernel_size=3, padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(filters=256, kernel_size=3, padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = SqueezeAndExcite(x)

    # Upsampling y capa final de predicción
    x = UpSampling2D((4, 4), interpolation="bilinear")(x)
    x = Conv2D(n_classes, 1, activation="sigmoid")(x)

    # Crear el modelo
    model = Model(inputs, x)
    return model
```
